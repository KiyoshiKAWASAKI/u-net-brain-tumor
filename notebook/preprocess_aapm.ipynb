```python
# Save dcm into images

import sys
import os
import SimpleITK as sitk
import numpy as np
import dicom
import matplotlib.pyplot as plt
from matplotlib import cm
import cv2
```

```python
file_path = "/Users/kiyoshi/all_research/all_medical_project/dataset/lung_cancer/train_valid/"
```

```python
for i in range (1, 37):
    if i < 10:
        dcm_folder_path = file_path + '000' + str(i) + '/all/dcm'
    else:
        dcm_folder_path = file_path + '00' + str(i) + '/all/dcm'

    dcm_files_path = os.listdir(dcm_folder_path)
    num_of_dcm = len(dcm_files_path)

    print "There are " + str(len(dcm_files_path)) + " dcm files in this folder."

    # Save the dcm files in each folder into a list
    list_of_dcm = []
    for j in range(num_of_dcm):
        list_of_dcm.append(dcm_files_path[j])

    for k in range(num_of_dcm):
        try:
            # Get raw data
            slices = dicom.read_file(dcm_folder_path + '/' + list_of_dcm[k])
            row = int(slices.Rows)
            column = int(slices.Columns)
            dimension = (row, column, num_of_dcm)
            spacing = (float(slices.PixelSpacing[0]), float(slices.PixelSpacing[1]), float(slices.SliceThickness))
            #print dimension

        except AttributeError:
            print "Cannot load the " + str(i+1) + "th file."

        try:
            # Calculate the dimension
            x = np.arange(0.0, (dimension[0]+1)*spacing[0], spacing[0])
            y = np.arange(0.0, (dimension[1]+1)*spacing[1], spacing[1])
            z = np.arange(0.0, (dimension[2]+1)*spacing[2], spacing[2])

            # Arrange pydicom data
            dicom_array = np.zeros(dimension, dtype = slices.pixel_array.dtype)
            dicom_array[:, :, k] = slices.pixel_array

            # Make new directory for the image
            save_base_path = '/Users/kiyoshi/all_research/all_medical_project/dataset/lung_cancer/images/'
            folder_index = str(i)
            save_path = save_base_path + folder_index
            if not os.path.exists(save_path):
                os.mkdir(save_path)
            else:
                pass

            # Save images into the respective folder
            """
            plt.figure(dpi=300)
            plt.axes().set_aspect('equal', 'datalim')
            plt.set_cmap(plt.gray())
            plt.pcolormesh(x, y, np.flipud(dicom_array[:, :, k]))
            plt.axis('off')
            plt.savefig(save_path+'/'+str(k)+'.png', bbox_inches='tight')
            print "Saving image " + str(k)
            plt.close()
            """
            norm = plt.Normalize(vmin = dicom_array[:, :, k].min(), vmax = dicom_array[:, :, k].max())
            plt.imsave(save_path + '/' + str(k)+'.jpg', plt.cm.bone(norm(dicom_array[:, :, k])))
            if k%10 == 0:
                print "Saving image " + str(k)
            plt.close()

        except TypeError:
            print "Cannot calculate the " + str(i+1) + "th file."

print "Transfer dcm into jpg finished!"

```
